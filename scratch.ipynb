```python
from bitbucket.client import Client
import getpass as gp
import pandas as pd
```

```python
def getCreds():
    un=gp.getpass('Username:')
    pw=gp.getpass('Pasword:')
    return (un,pw)

lol=getCreds()
```

```python
cli=Client(lol[0],lol[1],'aeaverification')
```

```python
names=[]
cont=True
i=1
while cont:
    d=cli.get_repositories(params={'pagelen':100,'page':i})
    if len(d['values'])==0:
        cont=False
    for j in range(len(d['values'])):
        print(d['values'][j]['name'])
        names=names+[(d['values'][j]['name'])]
        
    i+=1
```
